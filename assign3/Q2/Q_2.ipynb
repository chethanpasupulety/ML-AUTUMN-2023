### PART A: Probability

```python

import random
import matplotlib.pyplot as plt
import numpy as np


# function to get all the possible outcomes of a k-sided die
def calculate_sides(k):
    return list(range(1, k + 1))

# function to get the probabilities of all the possible outcomes of a k-sided die
def calculate_probabilities(k):
    probabilities = []
    for i in range(1, k + 1):
        if i != 1 and i != k:
            probability = 1 / (2 ** (i - 1))
        else:
            probability = 1 / (2 ** (k - 1))
        probabilities.append(probability)
    return probabilities


# function to simulate the outcome of 1 roll of the  k-sided die 
def simulate_k_sided_die(k):
    outcomes = calculate_sides(k)
    probabilities = calculate_probabilities(k)

    
    result = random.choices(outcomes, probabilities)[0]
    return result

# function to simulate the outcome of n rolls of the k-sided die and store the sum of the outcomes.
def simulate_rolls(num_sides, num_rolls):

    # call the simulate_k_sided_die function num_rolls times and store the results in a list
    simulation = [simulate_k_sided_die(num_sides) for _ in range(num_rolls)]
    
    sum = 0
    for i in range(len(simulation)):
    
        sum += simulation[i]
        
    return sum

# final function to simulate the outcome of n trials of the k-sided die and store the sum of the outcomes.
def final_simulation(num_sides, num_rolls, num_trials):
    
    # call the simulate_rolls function num_trials times and store the results in a list
    results = [simulate_rolls(num_sides, num_rolls) for _ in range(num_trials)]
    
    
    # Calculating the five-number summary
    minimum = min(results)
    maximum = max(results)
    median = sorted(results)[len(results)//2]
    q1 = sorted(results)[:len(results)//2][len(results)//4]
    q3 = sorted(results)[len(results)//2:][len(results)//4]

    print(f"Minimum: {minimum}")
    print(f"Q1: {q1}")
    print(f"Median: {median}")
    print(f"Q3: {q3}")
    print(f"Maximum: {maximum}")
    
    # Calculating the theoretical expected sum and mean of the simulated results
    # theoretical expected sum = sum of all possible outcomes * probability of each outcome
    #                          = expected value of a single roll * number of rolls
    #                         
    
    outcomes = calculate_sides(num_sides)
    probabilities = calculate_probabilities(num_sides)
    theoretical_expected_sum = 0
    for i in range(len(outcomes)):
        theoretical_expected_sum += outcomes[i] * probabilities[i]
        
    print(f"Theoretical Expected Sum: {theoretical_expected_sum*num_rolls}")

    mean_simulated = np.mean(results)
    print(f"Mean of Simulated Results: {mean_simulated}")
    
            
    # Plot the histogram with correct bins
    plt.hist(results, bins=range(minimum-1, maximum+1), align='left', rwidth=0.8)
    plt.xlabel(f'Sum of Die Rolls ({num_sides} times)')
    plt.ylabel('Frequency')
    plt.title('Frequency Distribution of Sum of Die Rolls')
    plt.show()

        

```

## Experimental Setup
1. **Bias of the Die:** The die is biased such that the probability \(P(i)\) of landing on face \(i\) follows the formula:
   \[
   P(i) = \begin{cases}
   \frac{1}{2^{i-1}}, & \text{for } 2 \leq i \leq K \\
   \frac{1}{2^{K-1}}, & \text{for } i = 1 \text{ or } i = K
   \end{cases}
   \]
2. **Simulation Parameters:**
   - \(K\) (Number of sides on the die)
   - \(num\_rolls\) (Number of times the die is rolled in a single trial)
   - \(num\_trials\) (Number of independent trials conducted)

### Defining the Model

As the given features are continuous, we will be using the Gaussian Naive Bayes model. The model is defined below.The explanation of the code is given in the comments.

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

class GaussianNaiveBayes:
    def fit(self, X_train, y_train):

        # class_means is a new numpy array with 2 rows and 57 columns
        # each row represents a class
        # each column has the mean of the corresponding feature
        self.class_means = X_train.groupby('Class').mean()

        # class_variances is a new numpy array with 2 rows and 57 columns
        # each row represents a class
        # each column has the variance of the corresponding feature
        self.class_variances = X_train.groupby('Class').var()

        # prior_y is a new numpy array with 2 elements
        # each element represents the prior probability of a class
        self.prior_y = X_train['Class'].value_counts(normalize=True).sort_index().values
        
    def gaussian(self, x, mean, variance):

        # gaussian function is used to calculate the likelihood of an unseen data point
        # returns the likelihood of x given mean and variance using the gaussian formula
        # these likelihoods are multiplied to the prior probability of the class to get the posterior probability of the class
        # the class with the highest posterior probability is the predicted class

        return (1 / (np.sqrt(2 * np.pi * variance))) * np.exp(-((x - mean) ** 2) / (2 * variance))
    
    def predict(self, X_test):

        # predicted_values is a new numpy array with the same number of rows as X_test
        # each element represents the predicted class of the corresponding row in X_test
        predicted_values = []

        # for each row in X_test we calculate the posterior probability of each class and choose the class with the highest probability
        for _, row in X_test.iterrows():

            # posterior_y is a copy of prior_y
            posterior_y = self.prior_y.copy()

            # for each class we calculate the likelihood of the row and multiply it to the prior probability of the class
            # here we only have 2 classes so we loop over [0, 1]
            for idx, class_label in enumerate([0, 1]):

                # for each feature in the row we calculate the likelihood of the feature given the class
                for feature_idx, feature_value in enumerate(row[:-1]):

                    # mean and variance are the mean and variance of the feature given the class
                    mean = self.class_means.loc[class_label, X_test.columns[feature_idx]]
                    variance = self.class_variances.loc[class_label, X_test.columns[feature_idx]]

                    # likelihood is the likelihood of the feature given the class calculated using the gaussian function
                    likelihood = self.gaussian(feature_value, mean, variance)

                    # we multiply the likelihood to the prior probability of the class to get the posterior probability of the class
                    # Here comes our key assumption of conditional independence used in the naive bayes classifier
                    posterior_y[idx] *= likelihood
            
            # we choose the class with the highest posterior probability
            predicted_values.append(np.argmax(posterior_y))
        return np.array(predicted_values)


# function to evaluate the model
def evaluate(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1score = f1_score(y_true, y_pred)
    return accuracy, precision, recall, f1score



# Example usage
model = GaussianNaiveBayes()
model.fit(X_train, y_train['Class'])
predicted_values = model.predict(X_test)
accuracy, precision, recall, f1score = evaluate(y_test['Class'], predicted_values)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1score)

    

    
```

## Why Did the Accuracy Increase After Applying Log Transformation?

The accuracy increased after applying the log transformation due to several reasons:

1. **Handling Skewed Distributions:**
   In the original dataset, features like word frequencies often exhibit skewed distributions where a few words occur frequently, while many words occur rarely. Log transformation compresses the scale of these features, making them more aligned with the underlying distribution. This transformation helps in capturing subtle patterns within the data more effectively.



2. **Equalizing the Impact:**
   Log transformation reduces the impact of outliers and extreme values. For word frequencies, certain words might have extremely high or low frequencies. Log transformation mitigates the effect of these extreme values, allowing the model to focus on overall patterns in the data rather than being heavily influenced by outliers.


3. **Normalization:**
   Log transformation acts as a form of normalization, making the feature distributions more symmetric and closer to a normal distribution. Many machine learning algorithms assume that features are normally distributed. Normalized features often lead to better model performance.


By transforming the features using the log function, the dataset became more suitable for the Gaussian Naive Bayes model, which assumes that features follow a normal distribution. As a result, the model could learn more accurate decision boundaries, leading to the observed increase in accuracy after applying the log transformation.
